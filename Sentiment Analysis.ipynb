```python
import nltk
import csv
import numpy as np
```

```python
negative = []
with open ('words_negative.csv','r') as file:
  re = csv.reader(file)
  for row in re:
    negative.append(row)
```

```python
negative
```

```python
positive = []
with open ('words_positive.csv','r') as file:
  re = csv.reader(file)
  for row in re:
    positive.append(row)
positive
```

```python
def senti(text):
  temp = []
  text_s = nltk.sent_tokenize(text)
  for s in text_s:
    n_count = 0
    p_count = 0
    s_word = nltk.word_tokenize(s)
    for i in s_word:
      for item in positive:
        if (i ==item[0]):
          p_count += 1
      
      for item in negative:
        if (i ==item[0]):
          n_count += 1
      
    if (p_count > 0 and n_count == 0):
      temp.append(1)
    elif (n_count%2 >0):
      temp.append(-1)
    elif (n_count%2==0 and n_count>0):
      temp.append(1)
    else:
      temp.append(0)

    return temp
```

```python
nltk.download('punkt')
```

```python
comments = []
with open ('reviews.csv','r') as file:
  re = csv.reader(file)
  for row in re:
    comments.append(row)
```

```python
for rev in comments:
  print('\n')
  print(np.average(senti(str(rev))))
  print(rev)
```
