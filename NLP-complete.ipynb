```python
from nltk.util import ngrams
```

```python
text2 = 'This dataset is taken from https://ai.stanford.edu/~amaas/data/sentiment/ and then preprocess to put all positive and negative reviews in the same file for training and testing. It help you to put more effort on algorithm instead of data collection'
```

```python
tokens = nltk.word_tokenize(text2)
```

```python
tokens
```

```python
bigrams = ngrams(tokens,2)
```

```python
for i in bigrams:
  print(i)
```

```python
fourgrams = ngrams(tokens,4)
for i in fourgrams:
  print(i)
```

```python
def n_grams(text,n):
  tokens = nltk.word_tokenize(text)
  grams = ngrams(tokens,n)
  return grams
```

```python
text3 = "This course introduces Natural Language Processing through the use of python and the Natural Language Tool Kit. Through a practical approach, you'll get hands on experience working with and analyzing text"
```

```python
for i in n_grams(text3,5):
  print(i)
```

#### Regular Expressions
 - ^ - Start of String
 - $ - End of String
 - . - Wild card Character
 - [chr] - Matches one of the Character
 - [a-m] - Matches one Character in the range
 - ? - The previous character occur 0 or 1 times
 - \* - The previous Character occures 0 or more times
 - \+ - The Previous Character occures 1 or more times
 - a|e - Matches one or the other
 - () - Parenthesis for grouping expressions
 - \ - Escape Character 

```python
import re
```

```python
alice = nltk.corpus.gutenberg.words('carroll-alice.txt')
```

```python
set([x for x in alice if re.search("^new",x)])
```

```python
set([x for x in alice if re.search("ful$",x)])
```

```python
set([x for x in alice if re.search("^..nn..$",x)])
```

```python
set([x for x in alice if re.search("[chr]at$",x)])
```

```python
set([x for x in alice if re.search("^[chr]at$",x)])
```

```python
set([x for x in alice if re.search("^.*nn.*$",x)])
```

```python
set([x for x in alice if re.search("^[aeiou]c.+y$",x)])
```
