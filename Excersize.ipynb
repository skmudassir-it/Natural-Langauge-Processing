```python
for s in inaugural.fileids():
  wds_t = len(inaugural.words(s))
  sent_t = len(inaugural.sents(s))
  print(wds_t/sent_t,s)
```

```python
alice = nltk.corpus.gutenberg.words('carroll-alice.txt')
```

```python
a_norm = [i.lower() for i in alice if i.isalpha()]
```

```python
nltk.download('averaged_perceptron_tagger')
nltk.download('universal_tagset')
```

```python
a_tag = nltk.pos_tag(a_norm,tagset='universal')
```

```python
a_cfd = nltk.ConditionalFreqDist(a_tag)
```

```python
a_cfd
```

```python
a_cfd['over']
```

# Example 4

```python
story = nltk.corpus.gutenberg.words('bryant-stories.txt')
```

```python
tags = nltk.pos_tag(story, tagset='universal')
```

```python
tags[:10]
```

```python
for ((word1, tag1),(word2, tag2),(word3, tag3)) in nltk.trigrams(tags):
  if tag1 =='NOUN' and tag2 == ' or' and tag3 =='NOUN':
    print (word1 +' '+ word2 +' '+ word3)
```
